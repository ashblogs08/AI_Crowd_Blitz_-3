```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from simpletransformers.classification import ClassificationModel
```

```python
del model
```

```python
# model = ClassificationModel('xlnet', 'xlnet-base-cased', num_labels=2, use_cuda=True)

model = ClassificationModel('roberta', 'roberta-base', num_labels=2, use_cuda=True,args={'fp16': False,'overwrite_output_dir': True,
})
```

```python
df_train = pd.read_csv('data/train.csv')
df_val = pd.read_csv('data/val.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
df_train['label'] = (df_train['label'] == 'real').astype(int)
df_val['label'] = (df_val['label'] == 'real').astype(int)
df_train.head()
```

```python
df_train = df_train.sample(frac=1)
df_train.head()
```

```python
model.train_model(df_train,output_dir="/tmp",args={'fp16': False,'overwrite_output_dir': True,
})
```

```python
# model.train_model(df_val,output_dir="/tmp",args={'fp16': False,'overwrite_output_dir': True,"learning_rate": 5e-6,
# })
```

```python
result, output, wrong = model.eval_model(df_val)
result
```

```python
y_val = list(df_val['label'])
y_pred = []
for i in output:
    y_pred.append(np.argmax(i))
```

```python
print(accuracy_score(y_val, y_pred))
print(precision_score(y_val, y_pred))
print(recall_score(y_val, y_pred))
print(f1_score(y_val, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_val,y_pred))
```

```python
print(accuracy_score(y_val, y_pred))
print(precision_score(y_val, y_pred))
print(recall_score(y_val, y_pred))
print(f1_score(y_val, y_pred))
```

```python
y_test = model.predict(df_test['text'])
```

```python
submission = []
for pred in y_test[0]:
    if pred == 1:
        submission.append('real')
    else:
        submission.append('fake')
```

```python
sub = pd.DataFrame(submission)
sub.to_csv('data/roberta2.csv', header=['label'], index=False)
```
